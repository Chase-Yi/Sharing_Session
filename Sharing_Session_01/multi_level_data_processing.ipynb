```python
# Import required libraries
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import numpy as np
```

```python
# Create raw data to simulate the Excel table
data = {
    "Unnamed: 0": [np.nan, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013],
    "Seattle": ["Bikes", 33323, 55342, 33234, 33252, 22332, 23331, 33532, 11001, 10221],
    "Unnamed: 2": ["Accessories", 13394, 19983, 18884, 19893, 18840, 18890, 18790, 11000, 9900],
    "Unnamed: 3": ["Miscellaneous", 4455, 5563, 3348, 2239, 2232, 4343, 3434, 8840, 8892],
    "Portland": ["Bikes", 33323, 55342, 33234, 33252, 22332, 23331, 33532, 11001, 10221],
    "Unnamed: 5": ["Accessories", 13394, 19983, 18884, 19893, 18840, 18890, 18790, 11000, 9900],
    "Unnamed: 6": ["Miscellaneous", 4455, 5563, 3348, 2239, 2232, 4343, 3434, 8840, 8892],
    "Vancouver": ["Bikes", 33323, 55342, 33234, 33252, 22332, 23331, 33532, 11001, 10221],
    "Unnamed: 8": ["Accessories", 13394, 19983, 18884, 19893, 18840, 18890, 18790, 11000, 9900],
    "Unnamed: 9": ["Miscellaneous", 4455, 5563, 3348, 2239, 2232, 4343, 3434, 8840, 8892],
}

# Convert the raw data into a DataFrame
df = pd.DataFrame(data)
df
```

```python
# Display basic information about the DataFrame
print("Basic DataFrame Info:")
df.info()
```

```python
# Make a copy of the original DataFrame and fill any NaN values with 'None'
new_df = df.replace(np.nan,'None').copy()
new_df
```

```python
# Transpose the DataFrame to switch rows and columns
# Reset the index to create a clean, tabular structure
new_df = new_df.transpose().reset_index()
new_df
```

```python
# Set the column names of the DataFrame to the values in the first row
new_df.columns = new_df.iloc[0]
new_df
```

```python
# Remove the first row (which is now used as the header) and reset the index, dropping the old index
new_df = new_df[1:].reset_index(drop=True)
new_df
```

```python
# Rename specific columns for clarity
new_df = new_df.rename(columns={'Unnamed: 0':'City', 'None':'Category'})
new_df
```

```python
# Identify rows where the 'City' column contains the value 'Unnamed' (indicating missing city names)
# Set these rows' 'City' value to None to explicitly mark them as missing
new_df.loc[new_df[new_df['City'].str.contains('Unnamed')].index, 'City'] = None
new_df
```

```python
# Forward fill (ffill) the 'City' column, which replaces missing values with the last valid value
new_df['City'] = new_df['City'].ffill()
new_df
```

```python
# Reshape the DataFrame from wide format to long format using the melt function
# 'id_vars' specifies which columns to keep as identifier variables (these will remain unchanged)
# 'var_name' assigns the name 'Year' to the column that will contain the previous column names
# 'value_name' assigns the name 'Value' to the column that will contain the corresponding values
new_melted_df = new_df.melt(id_vars=['City', 'Category'], var_name='Year', value_name='Value')
new_melted_df
```

```python
# Define a function to reorder and sort the DataFrame
def reorder_magic(special_order_list_1, special_order_list_2, df, col1, col2, col3):
    
    # Convert the specified columns to 'category' type for ordered sorting
    df[col1] = df[col1].astype('category')
    df[col2] = df[col2].astype('category')

    # Reorder categories in the specified order for the given columns
    df[col1] = df[col1].cat.reorder_categories(special_order_list_1)
    df[col2] = df[col2].cat.reorder_categories(special_order_list_2)

    # Sort the DataFrame by the specified columns
    df = df.sort_values(by=[col1, col2, col3]).reset_index(drop=True)

    return df
```

```python
# Apply the function to reorder and sort the DataFrame
new_melted_df = reorder_magic(
    ['Seattle', 'Portland', 'Vancouver'],  # Order for the 'City' column
    ['Bikes', 'Accessories', 'Miscellaneous'],  # Order for the 'Category' column
    new_melted_df,
    'City',  # First column to reorder
    'Category',  # Second column to reorder
    'Year'  # Column to sort by
)
new_melted_df
```
